## Question 5
 Calculate the Pearson Correlation coefficient of carat and price. Explain the results and discuss the strength of the correlation.

```python
# Import pearson library
from scipy.stats import pearsonr

# Convert required data into series to use pearson
carat_list = diamond_carat
price_list = diamond_price

# Apply the pearsonr() (using seatgeek pearson correlation test)
corr,_ = pearsonr(carat_list, price_list)
print("The Pearson Correlation coefficient of carat and price is %.2f" % corr)
```

The Pearson Correlation coefficient of carat and price is 0.92. This means that as the carat size of a diamond increases, its price generally increases as well, with carat weight being one of the most influential factors in determining the price. However, since the correlation is not a perfect value of 1, which indicates a perfect positive correlation, other diamond characteristics also play a role in price determination.

## Question 7
Using a 2-sample t-test**, determine if there is a statistical difference between the price of diamonds with a cut that is considered:
'Good' vs 'Very Good'
'Premium' vs 'Ideal'

```python
from scipy import stats
# Define the two sets of comparisons: 'Good' vs 'Very Good' and 'Premium' vs 'Ideal'
good_prices = diamond_df[diamond_df['cut'] == 'Good']['price']
very_good_prices = diamond_df[diamond_df['cut'] == 'Very Good']['price']
premium_prices = diamond_df[diamond_df['cut'] == 'Premium']['price']
ideal_prices = diamond_df[diamond_df['cut'] == 'Ideal']['price']

# Get the t-statistic and p-value
t_stat_good_vgood, p_val_good_vgood = stats.ttest_ind(good_prices, very_good_prices, equal_var = False)
t_stat_prem_ideal, p_val_prem_ideal = stats.ttest_ind(premium_prices, ideal_prices, equal_var = False)

alpha = 0.05

# Check results for Good vs Very Good
if p_val_good_vgood < alpha:
    print(f'The p-value for Good vs Very Good: {p_val_good_vgood:.2f} indicates there is a statistical difference between the price of diamonds '
          f'with a cut that is considered Good vs Very Good, t-statistic: {t_stat_good_vgood:.2f}')
else:
    print(f'T-Statistic for Good vs. Very Good: {t_stat_good_vgood:.2f}, P-Value: {p_val_good_vgood:.2f} indicates there is no statistical difference.')

# Check results for Premium vs Ideal
if p_val_prem_ideal < alpha:
    print(f'The p-value for Premium vs Ideal: {p_val_prem_ideal:.2f} indicates there is a significant difference between the price of diamonds '
          f'with a cut that is considered Premium vs Ideal, t-statistic: {t_stat_prem_ideal:.2f}')
else:
    print(f'T-Statistic for Premium vs. Ideal: {t_stat_prem_ideal:.2f}, P-Value: {p_val_prem_ideal:.2f} indicates there is no statistical difference between the price and cut.')
```

## Question 8
Many statistical test rely on the assumption that data is normally distributed. Indicate which distributions in question 6 were skewed, and research one possible technique* that can be used to handle skewed data. Then perform the following:
1. Demonstrate the selected technique to correct the skewed data.

In question 6, the distributions for all the qualities of cut were right-skewed. One possible technique that can be used to handle skewed data is log transformations (as the helpful tip states).

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Apply log transformation to correct skewed data (https://tariqueakhtar-39220.medium.com/log-transformation-and-visualizing-it-using-python-392cb4bcfc74)
diamond_df['log_price'] = np.log(diamond_df['price'])
```

2. Repeat the analysis in question 6 using the corrected data and explain the difference in the distribution with the previous results.

```python
cuts = ['Ideal', 'Premium', 'Very Good', 'Good', 'Fair']

for cut in cuts:
    # Get the prices for each cut in the list
    cut_prices = diamond_df[diamond_df['cut'] == cut]['log_price']
 
    # Visualize corrected distributions
    sns.histplot(cut_prices, kde=True)
    plt.title(f'Log-Transformed Price Distribution for {cut} Cut')
    plt.xlabel('Price')
    plt.ylabel('Frequency of the Price')
    plt.show()
```

The graphs display the log-transformed price distributions of diamonds by cut quality. Higher-quality cuts, like Ideal and Premium, show greater price variability with multiple peaks, while lower-quality cuts, such as Good and Fair, have more concentrated distributions, reflecting less variation in pricing. The distributions in question 6 highlights the right-skewness of the price distributions, noting that most diamonds are priced below 2,500, with fewer high-priced diamonds, particularly in higher-quality cuts. The log-transformed price data are now distributed more evenly, giving us a normal distribution as compared to the distributions in question 6, where they were right-skewed.

3. Repeat the statistical tests in question 7. Are there any changes in the statistical significance? Explain the results.

```python
# Define the two sets of comparisons: 'Good' vs 'Very Good' and 'Premium' vs 'Ideal'
log_good_prices = diamond_df[diamond_df['cut'] == 'Good']['log_price']
log_very_good_prices = diamond_df[diamond_df['cut'] == 'Very Good']['log_price']
log_premium_prices = diamond_df[diamond_df['cut'] == 'Premium']['log_price']
log_ideal_prices = diamond_df[diamond_df['cut'] == 'Ideal']['log_price']

# Re-run t-tests with log-transformed data
t_stat_log_good_vgood, p_val_log_good_vgood = stats.ttest_ind(log_good_prices, log_very_good_prices, equal_var=False)
t_stat_log_prem_ideal, p_val_log_prem_ideal = stats.ttest_ind(log_premium_prices, log_ideal_prices, equal_var=False)

# Set alpha
alpha = 0.05

# Good vs Very Good
if p_val_log_good_vgood < alpha:
    print(f'Log-Transformed Data: There is a significant difference between Good vs Very Good prices (p-value: {p_val_log_good_vgood:.2f}, t-statistic: {t_stat_log_good_vgood:.2f})')
else:
    print(f'Log-Transformed Data: No significant difference between Good vs Very Good prices (p-value: {p_val_log_good_vgood:.2f}, t-statistic: {t_stat_log_good_vgood:.2f})')

# Premium vs Ideal
if p_val_log_prem_ideal < alpha:
    print(f'Log-Transformed Data: There is a significant difference between Premium vs Ideal prices (p-value: {p_val_log_prem_ideal:.2f}, t-statistic: {t_stat_log_prem_ideal:.2f})')
else:
    print(f'Log-Transformed Data: No significant difference between Premium vs Ideal prices (p-value: {p_val_log_prem_ideal:.2f}, t-statistic: {t_stat_log_prem_ideal:.2f})')
```

The statistical significance of the price differences between diamond cuts changed notably after applying log transformation to correct for skewness. For the comparison between Good and Very Good cuts, the original analysis showed no significant difference (p-value: 0.41), but after log transformation, a significant difference emerged (p-value: 0.01), indicating that the transformation helped reveal underlying patterns in the data. In contrast, the comparison between Premium and Ideal cuts remained statistically significant in both analyses, with p-values of 0.00, suggesting a big difference in pricing. However, the log transformation increased the t-statistic from 24.92 to 28.11, indicating a stronger effect. Overall, the results demonstrate that log transformation can enhance the detection of significant differences by correcting for skewness, revealing insights that may not be apparent in the original data.
